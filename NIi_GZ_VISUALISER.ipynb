```python


# Install necessary libraries
!pip install nibabel
import nibabel as nib
import matplotlib.pyplot as plt

# Load NIfTI files from Google Drive
file_paths = [
    '/content/case0038_img.nii.gz',
    '/content/case0038_gt.nii.gz',
    '/content/case0038_pred.nii.gz',
]

# Create a subplot for each file
fig, axes = plt.subplots(1, len(file_paths), figsize=(15, 5))

for i, file_path in enumerate(file_paths):
    img = nib.load(file_path)
    img_data = img.get_fdata()

    # Extract the middle slice along the axial plane
    axial_slice_index = img_data.shape[2] // 2
    axial_slice = img_data[:, :, axial_slice_index]

    # Display the axial slice
    axes[i].imshow(axial_slice, cmap='gray')
    axes[i].set_title(f'Axial Slice - File {i+1}')

plt.show()
```
